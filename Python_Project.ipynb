# Drop Unamed Column

```python
df = df.drop("Unnamed: 0" , axis = 1)     #Drop Unamned Column
print(df.head())
```

# Change Weekly Study Hours Column

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("5 - 10","5-10")         #Replace Value
df.head()
```

# Gender Distriubution

```python
plt.figure(figsize = (5,5))                    #Take a fix size of graph in height and width form
ax = sns.countplot(data = df, x = "Gender")    #It is data and ax is variable
plt.title("GENDER DISTRIBUTION")
ax.bar_label(ax.containers[0])                 #Its show datalabels of the graph
plt.show()                                     #Its use show graph
```

# From the above chart we have analysed that - 
# The number of Females is more than the number of Males

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})  #agg is a aggreate function
print(gb)
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gb, annot = True)                 #Heatmap is chart type  #annot is = true show values
plt.title("RELATIONSHIP BETWEEN PARAENT'S EDUCATION AND STUDENT SCORES")
plt.show()
```

# From the above chart we have concluded that the education of the paraents have a good impact 

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gb1, annot = True)                 #Heatmap is chart type  #annot is = true show values
plt.title("RELATIONSHIP BETWEEN MARITAL STATUS AND STUDENT SCORES")
plt.show()
```

# From the above chart we have concluded that the marrital status of the parents have a negligible impact of eduction

# Distribution of Ethnic Groups

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df["EthnicGroup"] == "group E")].count()

v = ["GroupA", "GroupB", "GroupC", "GroupD", "GroupE"]
mylist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
plt.pie(mylist, labels = v, autopct = "%1.2f%%")
plt.title("DISTRIBUTION OF ETHNIC GROUPS")
plt.show()

```
